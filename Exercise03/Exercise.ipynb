a) [5%] Erstellen Sie einen Array mit 2000 Samplen aus einer 2D multivariaten Gauss Verteilung mit Mittelwert `np.array([0, 0])` und Covarianz `cov = np.array([[1, 0.], [0., 20]])`. 

Stellen Sie sicher, dass Sie Ihre Daten reproduzieren können, indem Sie sich informieren, was ein **Seed** ist. Setzen Sie diesen **Seed** auf die Zahl `101`.

```
Backup data:
data = np.load("backup_data_2a.npy")
```

````python
np.random.seed(101)
mean = np.array([0, 0])
cov = np.array([[1, 0.], [0., 20]])

# https://numpy.org/doc/stable/reference/random/generated/numpy.random.multivariate_normal.html
data = np.random.multivariate_normal(mean, cov, 2000)
#data = np.load("backup_data_2a.npy")
````

b) [10%] Definieren Sie eine Funktion, die Ihnen eine Rotationsmatrix abhängig von einem Rotationswinkel $\theta$ erstellt

````python
def get_rotation_matrix(theta):
    # Programmieren Sie diese Funktion
    # https://scipython.com/book/chapter-6-numpy/examples/creating-a-rotation-matrix-in-numpy/
    
    cos, sin = np.cos(theta), np.sin(theta)
    
    rotation_matrix = np.array(((cos, -sin), (sin, cos)))
    
    return rotation_matrix
````

````python
theta = np.radians(40)
r_matrix = get_rotation_matrix(theta)
data_rotated = data @ r_matrix
````

c) [25%] Erstellen Sie einen Plot, der möglichst ähnlich aussieht, wie der Beispielplot: 

Tipps: 
- Viele der hier verwendeten Funktionen wurden nicht in der Vorlesung oder im Tutorat behandelt - Google hilft!
- Sie können die Daten leicht in einem Scatterplot plotten, wenn Sie den Datenarray transponieren und dann mit `*` an die Plot-Funktion weitergeben: `ax.scatter(*data.T)`
- Das Kreuz, das den Mittelwert anzeigt, ist ein eigener Plot bestehend aus nur einem Punkt
- Die Transparanz in den Punkten und Histogrammen ist hilfreich, wenn man Dinge sichtbar machen will, die sich überlagern

````python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,6))

ax1.scatter(*data.T, color='blue', alpha=0.7, s=3, linewidths=2)
ax1.scatter(*data_rotated.T, color='orange', alpha=0.7, s=3, linewidths=2)
ax1.scatter(*np.mean(data, axis=0), marker='x', color='black', s=100, label='mean')

ax1.legend()
ax1.set(xlabel='x', ylabel='y')

ax2.hist(data, bins=200, alpha=0.7, histtype='stepfilled', label='Original data', color=['blue', 'orange'])
ax2.hist(data_rotated, bins=200, alpha=0.5, histtype='stepfilled', label='Rotated data', color=['orange']*2)
ax2.set(ylabel='count')
ax2.legend()

plt.tight_layout()
plt.show()
````

#### c) (5%) Bearbeiten Sie den Datensatz mithilfe von Pandas, sodass in der Spalte `Sex` nur eine binäre und konsistente Unterscheidung von männlich (m) und weiblich (f) zu finden ist. 

````python
# -1 men 1 women
df['Sex'] = df['Sex'].replace(r'1|Female|femaale|female|femal|femae|femalle', 'f', regex=True)
df['Sex'] = df['Sex'].replace(r'-1|male|mal|Male|mle|-f', 'm', regex=True)

print(df.to_string())
````

#### d) (5%) Bereinigen Sie die Spalte `Name`. Entfernen Sie Anreden wie *Mr., Mrs., Master*,... Verwenden Sie den in Klammern angegebenen Namen, falls vorhanden.

````python
import re

def clean_name(name):
    last_name = re.search('^.*\\,', name)

    remove_salut_regex = 'Mr\\.|Miss\\.|Master\\.|Mrs\\.\s*' 
    
    if not last_name:
        return re.sub(remove_salut_regex, '', name)
    # get match group and remove last char bc it contains ','
    last_name = last_name.group(0)[:-1]

    women_first_name = re.search('\\(.*\\)', name) 

    if women_first_name:
        # remove () from name
        women_first_name = women_first_name.group(0)[1:-1]

        women_name = f'{women_first_name} {last_name}'
        
        return re.sub(remove_salut_regex, '', women_name)

    men_first_name = re.search('\\,.*', name)

    if not men_first_name:
        return re.sub(remove_salut_regex, '', name)
    # remove ,
    men_first_name = men_first_name.group(0)[1:]

    men_name = f'{men_first_name} {last_name}'
    
    return re.sub(remove_salut_regex, '', men_name)

df['Name'] = df['Name'].map(clean_name)

print(df.to_string())
        
````
